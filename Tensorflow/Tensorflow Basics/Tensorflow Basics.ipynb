```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model = keras.Sequential([keras.layers.Dense(32,activation='relu' , input_shape = [1]),keras.layers.Dense(16,activation='relu' , input_shape = [1]),keras.layers.Dense(1)])
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
xs = np.random.random((10000, 1))*21
ys = xs**2
```

```python
model.fit(xs,ys,validation_split=0.2,epochs = 1000, batch_size = 256)
```

```python
print(model.predict([73]))
print(73**2)
```
